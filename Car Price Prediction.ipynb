## Linear regression model

### Declare the inputs and the targets

```python
targets = data_preprocessed['Log price']
inputs = data_preprocessed.drop(['Log price'], axis=1)
```

### Scale the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
```

```python
inputs_scaled
```

It is not usually recommended to standardize dummy variables

Scaling has no effect on the predictive power of dummies.

Once scaled, though, they lose all their dummy meaning

### Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

### Create the regression

```python
reg = LinearRegression()

reg.fit(x_train, y_train)
```

This reression is a log-linear regression as the dependent variable is the log of 'Price'

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y_train)', size=18)
plt.ylabel('Predictions (y_hat)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.distplot(y_train - y_hat)
plt.title('Residual PDF', size=18)
```

The residuals are the estimates of the errors

#### Normality and Homoscedasticity

* Normality
* Zero mean
* Homoscedasticity

Since there is more outliers to the left of the distplot, there are certain observations for which (y_train - y_hat) is much lower than the mean (a much higher price is predicted than is observed)

```python
reg.score(x_train, y_train)
```

### Finding the weights and bias

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

The bigger the weight, the bigger the impact

It doesn't make sense to compare the continuous variables with the dummies

Dummies are only compared to their respective benchmarks dummies for the respective discrete variabes

```python
data_cleaned['Brand'].unique() # Since Audi is the variable that got dropped, Audi is the benchmark
```

### Weight Interpretation
1. Countinuous Variables:
* A positive weight shows that as a feature increases in value, so do the log price, and price respectively
* A negative weight shows that as a feature increases in value, log price and price decreases

2. Dummy variables:
* A positive weight shows that the respective category (Brand) is more expensive than the benchmark (Audi)
* A negative weight shows that the respective category (Brand) is less expensive than the benchmark (Audi)

### Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test, alpha=0.2)
plt.xlabel('Targets (y_test)', size=18)
plt.ylabel('Predictions (y_hat_test)', size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

For higher prices we have a higher concentratin,, therefore our model is very good at predicting higher prices

However for lower ones they are much more scattered hence not doing as good as the former

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Predictions'])
df_pf.head()
```

To convert the log prices back to the actual prices we will take the exponential 

```python
df_pf['Target'] = np.exp(y_test)
# df_pf = df_pf.drop('Targets', axis=1)
df_pf.head()
```

```python
y_test = y_test.reset_index(drop=True) # You need to drop the index
y_test.head()
```

```python
df_pf['Target'] = np.exp(y_test)
df_pf
```

```python
df_pf['Residual'] = df_pf['Target'] - df_pf['Predictions']
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
```
